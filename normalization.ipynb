```python
import pandas as pd 
import numpy as np
```

```python
header_list = ["img_id", "class", "cx", "cy", "w", "h"]
train_labels = pd.read_csv("train/labels.txt", names=header_list, sep=" ")
val_labels = pd.read_csv("val/labels.txt", names=header_list, sep=" ")
```

```python
# Change class range from 1-7 to 0-6 in yolo format
train_labels['class'] = train_labels['class'] - 1
val_labels['class'] = val_labels['class'] - 1
```

```python
#Normalize training data 
w_img = 960
h_img = 540

train_labels['cx'] = train_labels['cx'] / w_img
train_labels['cy'] = train_labels['cy'] / h_img
train_labels['w'] = train_labels['w'] / w_img
train_labels['h'] = train_labels['h'] / h_img
```

```python
# Normalize validation data
val_labels['cx'] = val_labels['cx'] / w_img
val_labels['cy'] = val_labels['cy'] / h_img
val_labels['w'] = val_labels['w'] / w_img
val_labels['h'] = val_labels['h'] / h_img
```

```python
val_labels.head()
```

```python
#split labels per image

for i, g in train_labels.groupby(['img_id']):
    df = g[["class", "cx", "cy", "w", "h"]]
    #df['img_id'] = df['img_id'].astype(str).str.zfill(5)
    j = str(i).zfill(5) # to name the label file for each image same as the image name
    np.savetxt('/WAVE/users/unix/tpandey/yolov5/train/labels/'+j+'.txt', df.values, fmt="%.8g")
```

```python
#split validation labels per image
for i, j in val_labels.groupby(['img_id']):
    df_val = j[["class", "cx", "cy", "w", "h"]]
    k = str(i).zfill(5) # to name the label file for each image same as the image name
    np.savetxt('val/labels/'+k+'.txt', df_val.values, fmt="%.8g")
```
